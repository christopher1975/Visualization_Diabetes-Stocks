### Soal 2 (25 poin)
Tampilkan visualisasi dari korelasi (i) antar variabel dan (ii) antara variabel dan target, lalu jelaskan hasil korelasinya.

#### Code

```python
# type your code here

#korelasi dengan target
#variabel korelasi yang ada pada data_1
korelasi = data_1.corr()

#visualisasi heatmap dari variabel korelasi
plt.figure()
sns.set(font_scale=1.1)
fig,ax = plt.subplots(figsize=(15,10))
plt.title("Korelasi Data", fontsize=20, pad=20)
sns.heatmap(korelasi, annot=True, linewidths=0.01, square=True)
plt.show()
```

```python
# type your code here

#korelasi tanpa target
#variabel korelasi yang ada pada data_1 dan drop table outcome
korelasi_non = data_1.drop(["Outcome"], axis = 1)
korelasi_non = korelasi_non.corr()

#visualisasi heatmap dari variabel korelasi
plt.figure()
sns.set(font_scale=1.1)
fig,ax = plt.subplots(figsize=(15,10))
plt.title("Korelasi Data", fontsize=20, pad=20)
sns.heatmap(korelasi_non, annot=True, linewidths=0.01, square=True)
plt.show()
```

### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

```python
# type your code here

hangseng = pd.read_csv('Hang_Seng.csv')
hangseng_id = np.where(hangseng['Date'].str[:4] == '2019')
update_hangseng = hangseng.loc[hangseng_id]

nikkei = pd.read_csv('Nikkei.csv')
nikkei_id = np.where(nikkei['Date'].str[:4] == '2019')
update_nikkei = nikkei.loc[nikkei_id]

nasdaq = pd.read_csv('Nasdaq.csv')
nasdaq_id = np.where(nasdaq['Date'].str[:4] == '2019')
update_nasdaq = nasdaq.loc[nasdaq_id]
```

```python
hangseng.info()
```

```python
nikkei.info()
```

```python
nasdaq.info()
```

#### Code

```python
# type your code here

update_hangseng["Day_Perc_Change"] = update_hangseng["Adj Close"].pct_change()*100
update_nikkei["Day_Perc_Change"] = update_nikkei["Adj Close"].pct_change()*100
update_nasdaq["Day_Perc_Change"] = update_nasdaq["Adj Close"].pct_change()*100

update_hangseng.dropna(axis = 0, inplace = True)
update_nikkei.dropna(axis = 0, inplace = True)
update_nasdaq.dropna(axis = 0, inplace = True)
```

```python
update_hangseng.head(10)
```

```python
update_nikkei.head(10)
```

```python
update_nasdaq.head(10)
```

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

```python
update_hangseng["Trend"]= update_hangseng["Day_Perc_Change"].apply(lambda x:trend(x))
update_hangseng.head()
```

```python
update_nikkei["Trend"]= update_nikkei["Day_Perc_Change"].apply(lambda x:trend(x))
update_nikkei.head()
```

```python
update_nasdaq["Trend"]= update_nasdaq["Day_Perc_Change"].apply(lambda x:trend(x))
update_nasdaq.head()
```

```python
update_hangseng['Trend'].unique()
```

```python
update_nikkei['Trend'].unique()
```

```python
update_nasdaq['Trend'].unique()
```

```python
#visualisasi hangseng
hangseng_trend = update_hangseng.groupby('Trend')
labels_hangseng = hangseng_trend.count().index
explode_list_hangseng = [0.1,0,0,0,0,0]
pie_label_hangseng = sorted([i for i in update_hangseng.loc[:, 'Trend'].unique()])
plt.figure(figsize=(15,6))
plt.pie(hangseng_trend['Trend'].count(),
                            autopct='%1.1f%%',
                            startangle=180,    
                            shadow=True,      
                            labels =None,
                            pctdistance=1.1,
                            explode=explode_list_hangseng)
plt.title('Nilai daily percentage change perusahaan Hangseng tahun 2019',y=1.12)
plt.axis('equal') 
plt.legend(labels=labels_hangseng, loc='upper left') 
plt.show()
```

```python
#visualisasi nikkei
nikkei_trend = update_nikkei.groupby('Trend')
labels_nikkei = nikkei_trend.count().index
explode_list_nikkei = [0.1,0,0,0,0,0]
pie_label_nikkei = sorted([i for i in update_nikkei.loc[:, 'Trend'].unique()])
plt.figure(figsize=(15,6))
plt.pie(nikkei_trend['Trend'].count(),
                            autopct='%1.1f%%',
                            startangle=180,  
                            shadow=True,      
                            labels =None,
                            pctdistance=1.1,
                            explode=explode_list_nikkei)
plt.title('Nilai daily percentage change perusahaan Nikkei tahun 2019',y=1.12)
plt.axis('equal') 
plt.legend(labels=labels_nikkei, loc='upper left') 
plt.show()
```

```python
nasdaq_trend = update_nasdaq.groupby('Trend')
labels_nasdaq = nasdaq_trend.count().index
explode_list_nasdaq = [0.1,0.1,0,0,0,0,0]
pie_label_nasdaq = sorted([i for i in update_nasdaq.loc[:, 'Trend'].unique()])
plt.figure(figsize=(15,6))
plt.pie(nasdaq_trend['Trend'].count(),
                            autopct='%1.1f%%', 
                            startangle=180,     
                            shadow=True,       
                            labels =None,
                            pctdistance=1.1,
                            explode=explode_list_nasdaq)
plt.title('Nilai daily percentage change perusahaan Nasdaq tahun 2019',y=1.12)
plt.axis('equal') 
plt.legend(labels=labels_nasdaq, loc='upper left') 
plt.show()
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

```python
update_hangseng.info()
```

```python
update_nikkei.info()
```

#### Code

```python
# type your code here

# Adj close
gabungan = pd.concat([update_hangseng["Adj Close"], update_nikkei["Adj Close"], update_nasdaq["Adj Close"]], axis=1)
gabungan.columns = ["HangSeng", "Nikkei", "Nasdaq"]
gabungan.head()
```

```python
# drop missing value / nilai null 
gabungan.dropna(inplace = True, axis = 0)
gabungan.head()
```

```python
pct = gabungan.pct_change()*100
pct.dropna(inplace = True, how = "any", axis = 0)
pct.head()
```

```python
volatilitas = pct[['HangSeng', 'Nikkei', 'Nasdaq']].rolling(7).std()*np.sqrt(7)
volatilitas.plot(figsize = (20, 10))
```
